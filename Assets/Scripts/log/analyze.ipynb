```python
with open(r"C:\Users\shigetomo\Desktop\UnityProject\InfiltrationTacticsDragonHunter\Assets\Scripts\log\eska_attack.log", "r") as f:
	logtext = f.readlines()

print(logtext)
```

```python
unit_damage = []
for log in logtext:
	aggregated_data = {}
	damage = 0
	for unit in log.split("/"):
		if not "," in unit:
			damage = int(unit.strip())
			continue

		name, number = unit.split(",")
		name = name.strip()
		number = number.strip()

		number = int(number)

		if name in aggregated_data:
			aggregated_data[name] += number
		else:
			aggregated_data[name] = number
	unit_damage.append([damage, aggregated_data])
```

```python
sorted(unit_damage, key=lambda x: x[0], reverse=True)
```
